## Sparse Matrix

```python
import numpy as np
import matplotlib.pyplot as plt
```

A sparse matrix is a matrix holding many zeros. 

Here, we've randomly put a random number (different of 0) in `int(n/16)` cases, where n is the size of the matrix.

```python
n=100
tab = np.zeros((int(n/16)))
matrice = np.zeros((n,n))
for i in range(int(n/16)):
    tab[i] = np.random.randint(1,n)
    x = np.random.randint(1,n)
    y = np.random.randint(1,n)
    matrice[x,y] = tab[i]
plt.imshow(matrice)
plt.colorbar()
```
